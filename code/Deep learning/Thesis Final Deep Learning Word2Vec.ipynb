```python
from bangla_stemmer.stemmer import stemmer
def stemmerbangla(data):
  wordlist = data.split(" ")
  stmr = stemmer.BanglaStemmer()
  stm = stmr.stem(wordlist)
  return " ".join(stm)
```

```python
import re
def onlybangla(data):
  bangla=data.replace('\r', ' ').replace('\n', ' ')
  bangla = "".join(i for i in bangla if i in [" "] or 2432 <= ord(i) <= 2559 or ord(i)== 32)
  return " ".join(bangla.split())
```

```python
a_file = open(os.getcwd()+"/stopwords-bn.txt", "r")

bn_stopwords = []
for line in a_file:
  stripped_line = line.strip()
  bn_stopwords.append(stripped_line)

a_file.close()
```

```python
def remove_stop_word(text):
  words = [word for word in text.split(" ") if word not in bn_stopwords]
  new_text = " ".join(words)
  return new_text
```

```python
from tensorflow.keras.layers import TextVectorization

vectorizer_text = TextVectorization(max_tokens=20000, output_sequence_length=200)
text_ds = tf.data.Dataset.from_tensor_slices(X_train['text']).batch(128)
vectorizer_text.adapt(text_ds)

vectorizer_comments = TextVectorization(max_tokens=20000, output_sequence_length=200)
comments_ds = tf.data.Dataset.from_tensor_slices(X_train['comments']).batch(128)
vectorizer_comments.adapt(comments_ds)
```

```python
import keras
```

```python
voc_text = vectorizer_text.get_vocabulary()
text_word_index = dict(zip(voc_text, range(len(voc_text))))
voc_comments = vectorizer_comments.get_vocabulary()
comments_word_index = dict(zip(voc_comments, range(len(voc_comments))))
```

```python
text_num_tokens = len(voc_text) + 2
text_embedding_dim = 100
text_hits = 0
text_misses = 0

# Prepare embedding matrix
text_embedding_matrix = np.zeros((text_num_tokens, text_embedding_dim))
for word, i in text_word_index.items():
    if word in wordvec.wv:
        text_embedding_vector = wordvec.wv[word]
    else:
        text_embedding_vector=None
    if text_embedding_vector is not None:
        # Words not found in embedding index will be all-zeros.
        # This includes the representation for "padding" and "OOV"
        text_embedding_matrix[i] = text_embedding_vector
        text_hits += 1
    else:
        text_misses += 1
print("Converted %d words (%d misses)" % (text_hits, text_misses))
```

```python
comments_num_tokens = len(voc_comments) + 2
comments_embedding_dim = 100
comments_hits = 0
comments_misses = 0

# Prepare embedding matrix
comments_embedding_matrix = np.zeros((comments_num_tokens, comments_embedding_dim))
for word, i in comments_word_index.items():
    if word in wordvec.wv:
        comments_embedding_vector = wordvec.wv[word]
    else:
        comments_embedding_vector=None
    if comments_embedding_vector is not None:
        # Words not found in embedding index will be all-zeros.
        # This includes the representation for "padding" and "OOV"
        comments_embedding_matrix[i] = text_embedding_vector
        comments_hits += 1
    else:
        comments_misses += 1
print("Converted %d words (%d misses)" % (comments_hits, comments_misses))
```

```python
x_train_text = vectorizer_text(np.array([[s] for s in X_train['text']])).numpy()
x_train_comments = vectorizer_comments(np.array([[s] for s in X_train['comments']])).numpy()

x_test_text = vectorizer_text(np.array([[s] for s in X_test['text']])).numpy()
x_test_comments = vectorizer_comments(np.array([[s] for s in X_test['comments']])).numpy()

x_val_text = vectorizer_text(np.array([[s] for s in X_val['text']])).numpy()
x_val_comments = vectorizer_comments(np.array([[s] for s in X_val['comments']])).numpy()
y_train = np.array(y_train)
y_test = np.array(y_test)
y_val = np.array(y_val)
```

```python
from tensorflow.keras.layers import Embedding

text_embedding_layer = Embedding(
    text_num_tokens,
    text_embedding_dim,
    embeddings_initializer=keras.initializers.Constant(text_embedding_matrix),
    trainable=False,
)
comments_embedding_layer = Embedding(
    comments_num_tokens,
    comments_embedding_dim,
    embeddings_initializer=keras.initializers.Constant(comments_embedding_matrix),
    trainable=False,
)
```

```python
import matplotlib.pyplot as plt

# Plot utility
def plot_graphs(history, string):
  plt.plot(history.history[string])
  plt.plot(history.history['val_'+string])
  plt.xlabel("Epochs")
  plt.ylabel(string)
  plt.legend([string, 'val_'+string])
  plt.show()
```

```python
from tensorflow.keras import layers

text_sequences_input = keras.Input(shape=(None,), dtype="int64")
text_embedded_sequences = text_embedding_layer(text_sequences_input)

comments_sequences_input = keras.Input(shape=(None,), dtype="int64")
comments_embedded_sequences = comments_embedding_layer(comments_sequences_input)

merged = tf.keras.layers.Concatenate(axis=1)([text_embedded_sequences, comments_embedded_sequences])
merged = layers.Conv1D(128, 5, activation="relu")(merged)
merged = layers.MaxPooling1D(2)(merged)
merged = layers.Conv1D(64, 5, activation="relu")(merged)
merged = layers.GlobalMaxPooling1D()(merged)
merged = layers.Dense(128, activation="relu")(merged)
merged = layers.Dropout(0.5)(merged)
merged = layers.Dense(64, activation="relu")(merged)
merged = layers.Dropout(0.5)(merged)
preds = layers.Dense(1, activation="sigmoid")(merged)

modecnn = keras.Model(inputs=[text_sequences_input, comments_sequences_input], outputs=[preds])
modecnn.summary()
```

```python
tf.keras.utils.plot_model(model, "my_first_model_with_shape_info.png", show_shapes=True)
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
history=model.fit(x = [x_train_text,x_train_comments],y = y_train, batch_size=128, epochs=20, validation_data=([x_val_text,x_val_comments], y_val),callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)])
```

```python
# Plot the accuracy and loss
plot_graphs(history, "accuracy")
plot_graphs(history, "loss")
```

```python
train_cnn_results = model.evaluate(x=[x_train_text,x_train_comments], y=np.asarray(y_train), verbose=0, batch_size=128)
test_cnn_results = model.evaluate(x=[x_test_text,x_test_comments], y=y_test, verbose=0, batch_size=128)
print(f'Train accuracy: {train_cnn_results[1]*100:0.2f}')
print(f'Test accuracy: {test_cnn_results[1]*100:0.2f}')
```
