```python
#for data analysis and modeling
import tensorflow as tf
from tensorflow.keras.layers import LSTM, GRU, Dense, Embedding, Dropout
from tensorflow.keras.preprocessing import text, sequence 
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
#for text cleaning
import string
#for visualization
import matplotlib.pyplot as plt
import os

import json
```

```python
rumor_path=os.getcwd()+'/rumor'
non_rumor_path=os.getcwd()+'/non_rumor'
```

```python
rumorList = list()
for (dirpath, dirnames, filenames) in os.walk(rumor_path):
    rumorList += [os.path.join(dirpath, file) for file in filenames if file.endswith('.json')]
len(rumorList)
```

```python
non_rumorList = list()
for (dirpath, dirnames, filenames) in os.walk(non_rumor_path):
    non_rumorList += [os.path.join(dirpath, file) for file in filenames if file.endswith('.json')]
len(non_rumorList)
```

```python
from bangla_stemmer.stemmer import stemmer
def stemmerbangla(data):
  wordlist = data.split(" ")
  stmr = stemmer.BanglaStemmer()
  stm = stmr.stem(wordlist)
  return " ".join(stm)
```

```python
import re
def onlybangla(data):
  bangla=data.replace('\r', ' ').replace('\n', ' ')
  bangla = "".join(i for i in bangla if i in [" "] or 2432 <= ord(i) <= 2559 or ord(i)== 32)
  return " ".join(bangla.split())
```

```python
a_file = open(os.getcwd()+"/stopwords-bn.txt", "r")

bn_stopwords = []
for line in a_file:
  stripped_line = line.strip()
  bn_stopwords.append(stripped_line)

a_file.close()
```

```python
def remove_stop_word(text):
  words = [word for word in text.split(" ") if word not in bn_stopwords]
  new_text = " ".join(words)
  return new_text
```

```python
def rawtodf(listdata):
  listofdata=[]
  commentscount=0
  for datas in listdata:
    with open(datas, encoding='utf-8') as fh:
      data = json.load(fh)
      if data['text']!="" and len(data['comments'])!=0:
        commentscount+=len(data['comments'])
        listofdata.append([stemmerbangla(remove_stop_word(onlybangla(data['text']))),stemmerbangla(remove_stop_word(onlybangla("".join(data['comments'])))),data['label']])
  return pd.DataFrame(listofdata, columns = ["text","comments","label"]),commentscount
      
    #listofRumorandNonrumor.append([data['text'],data['sharecount'],(lambda: 1, lambda: 0)[data['post_shared_link']==""](),data['reactions']['likes']/data['reactioncount'],data['reactions']['loves']/data['reactioncount'],data['reactions']['cares']/data['reactioncount'],data['reactions']['hahas']/data['reactioncount'],data['reactions']['wows']/data['reactioncount'],data['reactions']['sads']/data['reactioncount'],data['reactions']['angries']/data['reactioncount'],data['commentcount'],data['comments'][0]['comment'],data['posted_by']['isVerify'],data['isRumor']])
```

```python
df=rumor.append(non_rumor)
df=df.drop_duplicates(subset=['text'], inplace=False)
```

```python
df_shuffle = df.sample(frac=1)
y=df_shuffle['label']
X=df_shuffle.drop(['label'], axis=1, inplace=False)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_dev, y_train, y_dev = train_test_split(X, y, test_size=0.30, random_state=42,stratify=y)
X_test, X_val, y_test, y_val = train_test_split(X_dev, y_dev, test_size=0.50, random_state=42,stratify=y_dev)
```

```python
train_text=X_train['text'].to_numpy()
test_text=X_test['text'].to_numpy()
val_text=X_val['text'].to_numpy()
```

```python
# Vocabulary size of the tokenizer
vocab_size = 10000
# Text Maximum length of the padded sequences
max_length=200
#Dimension
embedding_dim = 300

trunc_type='post'
padding_type='post'
oov_tok = "<OOV>"
```

```python
text_tokenizer = text.Tokenizer(oov_token=oov_tok,num_words=vocab_size)
text_tokenizer.fit_on_texts(train_text)

# Generate and pad the training sequences
training_text_sequences = text_tokenizer.texts_to_sequences(train_text)
training_text_padded = sequence.pad_sequences(training_text_sequences, maxlen=max_length, padding=padding_type)

# Generate and pad the testing sequences
testing_text_sequences = text_tokenizer.texts_to_sequences(test_text)
testing_text_padded = sequence.pad_sequences(testing_text_sequences, maxlen=max_length, padding=padding_type, truncating=trunc_type)

# Generate and pad the val sequences
validation_text_sequences = text_tokenizer.texts_to_sequences(val_text)
validation_text_padded = sequence.pad_sequences(validation_text_sequences, maxlen=max_length, padding=padding_type, truncating=trunc_type)
text_word_index = text_tokenizer.word_index
```

```python
train_comments=X_train['comments'].to_numpy()
test_comments=X_test['comments'].to_numpy()
val_comments=X_val['comments'].to_numpy()
```

```python
comments_tokenizer = text.Tokenizer(oov_token="<OOV>",num_words=vocab_size)
comments_tokenizer.fit_on_texts(train_comments)

# Generate and pad the training sequences
training_comments_sequences = comments_tokenizer.texts_to_sequences(train_comments)
training_comments_padded = sequence.pad_sequences(training_comments_sequences, maxlen=max_length, padding=padding_type)

# Generate and pad the testing sequences
testing_comments_sequences = comments_tokenizer.texts_to_sequences(test_comments)
testing_comments_padded = sequence.pad_sequences(testing_comments_sequences, maxlen=max_length, padding=padding_type, truncating=trunc_type)

# Generate and pad the val sequences
validation_comments_sequences = comments_tokenizer.texts_to_sequences(val_comments)
validation_comments_padded = sequence.pad_sequences(validation_comments_sequences, maxlen=max_length, padding=padding_type, truncating=trunc_type)
comments_word_index = comments_tokenizer.word_index
```

```python
training_labels=np.array(y_train)
testing_labels=np.array(y_test)
validation_labels=np.array(y_val)
```

```python
%%time
embedding_vectors = {}
# with open('/kaggle/input/glove6b100d/glove.6B.100d.txt','r',encoding='utf-8') as file:
with open(os.getcwd()+'/glove/bn_glove.39M.300d.txt','r',encoding='utf-8') as file:
    for row in file:
        values = row.split(' ')
        word = values[0]
        weights = np.asarray([float(val) for val in values[1:]])
        embedding_vectors[word] = weights
print(f"Size of vocabulary in GloVe: {len(embedding_vectors)}")   
```

```python
#initialize the embedding_matrix with zeros
emb_dim = 300
text_vocab_len = len(text_word_index)
text_embedding_matrix = np.zeros((vocab_len, emb_dim))
text_oov_count = 0
text_oov_words = []
for word, idx in text_word_index.items():
    if idx < vocab_len:
        embedding_vector = embedding_vectors.get(word)
        if embedding_vector is not None:
            text_embedding_matrix[idx] = embedding_vector
        else:
            text_oov_count += 1 
            text_oov_words.append(word)
#print some of the out of vocabulary words
print(f'Some out of valubulary words: {text_oov_words[0:5]}')
```

```python
#initialize the embedding_matrix with zeros
emb_dim = 300
comments_vocab_len = len(comments_word_index)
comments_embedding_matrix = np.zeros((comments_vocab_len, emb_dim))
comments_oov_count = 0
comments_oov_words = []
for word, idx in comments_word_index.items():
    if idx < vocab_len:
        embedding_vector = embedding_vectors.get(word)
        if embedding_vector is not None:
            comments_embedding_matrix[idx] = embedding_vector
        else:
            comments_oov_count += 1 
            comments_oov_words.append(word)
#print some of the out of vocabulary words
print(f'Some out of valubulary words: {comments_oov_words[0:5]}')
```

```python
import matplotlib.pyplot as plt

# Plot utility
def plot_graphs(history, string):
  plt.plot(history.history[string])
  plt.plot(history.history['val_'+string])
  plt.xlabel("Epochs")
  plt.ylabel(string)
  plt.legend([string, 'val_'+string])
  plt.show()
  
```

```python
text_inp = tf.keras.layers.Input(max_length)
text_emb = tf.keras.layers.Embedding(text_vocab_len, embedding_dim, input_length=max_length, trainable = False, weights=[text_embedding_matrix])(text_inp)
comments_inp = tf.keras.layers.Input(max_length)
comments_emb = tf.keras.layers.Embedding(comments_vocab_len, embedding_dim, input_length=max_length,trainable = False, weights=[comments_embedding_matrix])(comments_inp)

# concatenate them to makes a single vector per sample
merged = tf.keras.layers.Concatenate(axis=1)([text_emb, comments_emb])
merged = tf.keras.layers.Conv1D(128, 5, activation="relu")(merged)
merged = tf.keras.layers.MaxPooling1D(2)(merged)
merged = tf.keras.layers.Conv1D(64, 5, activation="relu")(merged)
merged = tf.keras.layers.GlobalMaxPooling1D()(merged)
merged = tf.keras.layers.Dense(128, activation="relu")(merged)
merged = tf.keras.layers.Dropout(0.5)(merged)
merged = tf.keras.layers.Dense(64, activation="relu")(merged)
merged = tf.keras.layers.Dropout(0.5)(merged)
out = tf.keras.layers.Dense(1, activation='sigmoid')(merged)
modelcnn = tf.keras.Model(inputs=[text_inp, comments_inp], outputs=[out])
modelcnn.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
modelcnn.summary()
```

```python
historycnn=modelcnn.fit(x = [training_text_padded,training_comments_padded], y = training_labels, epochs = 20, validation_data = ([validation_text_padded,validation_comments_padded],validation_labels),callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)])
```
